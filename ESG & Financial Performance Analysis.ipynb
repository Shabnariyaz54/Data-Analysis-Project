<h2 style="font-weight: bold;color: #DC143C"> DATA CLEANING & PRE-PROCESSING</h2>

<h4 style="font-weight: bold;color:green"> MISSING VALUES</h4>

```python
# Check for missing values in important numeric columns and replace with median

numeric_cols = ["Revenue","ProfitMargin","MarketCap","GrowthRate","ESG_Overall"]
for col in numeric_cols:
    if df[col].isna().sum()>0:
        df[col] = pd.to_numeric(df[col], errors='coerce')
        df[col].fillna(df[col].median(),inplace=True)     

# If ESG_Overall is missing, calculate it as the average of Environment, Social and Governance Score

ESG_Overall_Missing = df["ESG_Overall"].isna()
df.loc[ESG_Overall_Missing, "ESG_Overall"] = df.loc[ESG_Overall_Missing, ["ESG_Environmental","ESG_Social","ESG_Governance"]].mean(axis = 1)

```

<h4 style="font-weight: bold;color:green"> REMOVE DUPLICATES</h4>

```python
# Remove duplicates for the same company in the same year

df.drop_duplicates(subset = ["CompanyID","Year"],keep='first',inplace=True)


```

<h4 style="font-weight: bold;color:green"> EXTRA SPACES</h4>

```python
# Checking for extra spaces in each column

String_columns = ['CompanyName',
 'Industry',
 'Region',
 'Year',
 'Revenue',
 'ProfitMargin',
 'MarketCap',
 'GrowthRate',
 'ESG_Overall',
 'ESG_Environmental',
 'ESG_Social',
 'ESG_Governance',
 'CarbonEmissions',
 'WaterUsage',
 'EnergyConsumption']
for column in String_columns:
    if column in df.columns:
        has_spaces = df[column].astype(str).str.contains(r'^\s|\s$|\s{2,}').any()
        print(f"{column}: {has_spaces}")
```

```python
# Removing Extra space

extra_space_columns = [
 'Industry',
 'Region',
  ]

for column in extra_space_columns:
    if column in df.columns:
        df[column] = df[column].str.replace(r'\s+', ' ', regex=True).str.strip()
       
# Check removed space
for column in extra_space_columns:
    if column in df.columns:
        has_spaces = df[column].astype(str).str.contains(r'^\s|\s$|\s{2,}').any()
        print(f"{column}: {has_spaces}")
```

<h4 style="font-weight: bold;color:green"> CONVERT TO PROPER CASE</h4>

```python
# List of columns to convert to proper case
text_columns = [
 'Company_Name',
 'Industry',
 'Region',
]

for col in text_columns:
    if col in df.columns:
        df[col] = df[col].str.title()
        print(f"Converted {col} to proper case")
```

<h4 style="font-weight: bold;color:green"> CONVERT TO STRING </h4>

```python
# List of columns to convert to string
string_columns = ['CompanyName',
 'Industry',
 'Region'
                 ]

for col in string_columns:
    if col in df.columns:
        df[col] = df[col].astype('string')
        print(f"Converted {col} to string dtype")
```

<h4 style="font-weight: bold;color:green"> FORMATTING & STANDARDIZATON</h4>

```python
# Formatting & Standardization 

df["CompanyName"] = df["CompanyName"].str.strip()
df["Industry"] = df["Industry"].str.strip().str.title()
df["Region"] = df["Region"].str.strip().str.title()
df["Year"] = df["Year"].astype(int)

```

<h4 style="font-weight: bold;color:green"> AGGREGATING </h4>

```python
# Aggregating

aggregated = df.groupby('Industry').agg({
    'Revenue': 'mean',
    'ESG_Overall': 'mean'
})

print(aggregated)
```

<h4 style="font-weight: bold;color:green"> DERIVED FEATURES</h4>

```python
# Derived Features

df['Log_Revenue'] = np.log(df['Revenue'].replace(0, 1))
df['Log_MarketCap'] = np.log(df['MarketCap'].replace(0, 1))

# Profit - calculates absolute profit

df['Profit'] = df['Revenue'] * (df['ProfitMargin'] / 100)
pd.set_option('display.float_format','{:.2f}'.format)

df.head(5)
```

<h3 style="font-weight: bold;color: black">ðŸ“Š ESG Pillars : Feature importance in predicting revenue and profit</h3>

```python

esg_cols = ['ESG_Environmental', 'ESG_Social', 'ESG_Governance']

# Features and targets
X = df[esg_cols]
y_revenue = df['Revenue']
y_profit = df['Profit']

# Random Forest for Revenue
rf_rev = RandomForestRegressor(n_estimators=100, random_state=42)
rf_rev.fit(X, y_revenue)
rev_importance = pd.Series(rf_rev.feature_importances_, index=esg_cols)

# Random Forest for Profit
rf_profit = RandomForestRegressor(n_estimators=100, random_state=42)
rf_profit.fit(X, y_profit)
profit_importance = pd.Series(rf_profit.feature_importances_, index=esg_cols)

# Plotting side-by-side bar chart
bar_width = 0.35
indices = np.arange(len(esg_cols))

plt.figure(figsize=(10,6))
plt.bar(indices, rev_importance, width=bar_width, color='skyblue', label='Revenue')
plt.bar(indices + bar_width, profit_importance, width=bar_width, color='salmon', label='Profit')

# Add value labels on top of bars
for i in range(len(esg_cols)):
    plt.text(i, rev_importance[i]+0.01, f"{rev_importance[i]:.2f}", ha='center')
    plt.text(i + bar_width, profit_importance[i]+0.01, f"{profit_importance[i]:.2f}", ha='center')

plt.xticks(indices + bar_width/2, esg_cols)
plt.ylabel('Feature Importance')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

<h4 style="font-weight: bold;color: #DC143C">ðŸ” INSIGHT : </h4>

<div style="font-family: 'verdana'; line-height: 1.6; "><p style="font-size: 16px; text-align: justify;">
This chart reveals that ESG_Social is the most influential pillar in predicting both revenue and profit, outperforming Environmental and Governance dimensions. This suggests that companies with strong social practicesâ€”such as employee well-being, community engagement, and diversityâ€”may be better positioned for financial success. While Environmental and Governance factors also contribute, their impact is comparatively moderate.

</div>
